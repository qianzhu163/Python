```python
import pandas as pd
```

```python
import matplotlib as plt
```

```python
df=pd.read_csv('tradeHistory_test.csv',usecols=[0,2,5,6,9],encoding='gbk')
df
```

```python
def multiple(x):
    if '1810' in x:
        return 10
    else:
        return 100
df['multiple']=df['证券代码'].apply(multiple)
```

```python
df=df[df['证券代码']!='IF1812']
```

```python
df1=pd.DataFrame(df.groupby(['证券代码','交易日期','结算价','multiple'])['成交数量','成交金额'].sum(),columns=['成交数量','成交金额'])
```

```python
df1['hedgenumber']=df1.groupby(['证券代码'])['成交数量'].cumsum()*(-1)
```

```python
df1=df1.reset_index()
```

```python
df1['hedge']=df1['multiple']*df1['hedgenumber']*df1['结算价']
```

```python
df1['yesterdayhedge']=df1.groupby(['证券代码'])['hedge'].shift(1)*(-1)
```

```python
df1['日净收益']=(df1['成交金额']+df1['hedge']+df1['yesterdayhedge'])*(-1)
```

```python
df_pic=pd.DataFrame(df1.groupby(['交易日期'])['日净收益'].sum(),columns=['日净收益'])
df_pic['累计净收益']=df_pic['日净收益'].cumsum()
print(df_pic)
```

```python
df_pic.to_csv('futures_2.csv',encoding='gbk')
```
